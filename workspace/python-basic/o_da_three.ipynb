```python
# subway-data-1.csv 파일의 데이터를 읽어서 리스트로 구성
# csv 모듈 사용
# 숫자는 숫자형으로 변경

import csv

rows = []
with open('data-files/subway-data-1.csv', 'rt', encoding='utf-8') as f:
    reader = csv.reader(f)
    
    columns = next(reader) # 현재 행 읽고 다음 행으로 이동

    for row in reader: # 반복해서 row = next(reader)를 실행
        # print(row)
        for idx in range(0+4, len(row)-1):
            row[idx] = int(row[idx].replace(',', ''))
        rows.append(row)

rows
```
