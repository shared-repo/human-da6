```python
# 데이터 분석 모듈 준비

import numpy as np
import pandas as pd
```

```python
# seoul-temperature.csv 파일의 데이터를 읽어서 데이터프레임 만들기

seoul_temp_df = pd.read_csv('data-files/seoul-temperature.csv')
```

```python
# 데이터프레임 정보 확인

seoul_temp_df.info() # 데이터프레임의 정보를 요약해서 표시
print(seoul_temp_df.shape)
```

```python
# 앞 부분 일부 데이터를 표시
seoul_temp_df.head(n=3) # n : 표시할 행 갯수 (기본 값은 5)
```

```python
# 뒷 부분 일부 데이터 표시
seoul_temp_df.tail(n=7)
```

```python
# 최고 기온 날짜 찾기 1
# 최고기온 컬럼 기준 내림 차순 정렬 -> 앞부분 데이터 확인

sorted_df = seoul_temp_df.sort_values("최고기온(℃)", ascending=False) 

sorted_df.head()
```

```python
# 최고 기온 날짜 찾기 2
# 최고기온 찾기 -> 행 중에서 찾은 최고기온 값과 같은 행 찾기

# 최고기온 찾기
max_temp = seoul_temp_df['최고기온(℃)'].max()

max_mask = seoul_temp_df['최고기온(℃)'] == max_temp # max_temp와 같은 값의 위치만 True, 나머지는 False

seoul_temp_df[max_mask]
```
